```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('./0816_NN_Wordcloud.csv', index_col=0)
df.head()
```

```python
dic_freq = df.to_dict()['Pubmed']
dic_freq
```

```python
dic_freq = df.to_dict()['Pubmed']
keywords = dic_freq

wordcloud = WordCloud(
    width = 2000,
    height = 2000,
    background_color="white",
    max_words=1000,
    colormap='Paired'
)

wordcloud = wordcloud.generate_from_frequencies(keywords)
array = wordcloud.to_array()
fig = plt.figure(figsize=(10,4))
plt.imshow(array, interpolation="bilinear")
plt.axis("off")
plt.show()
```
